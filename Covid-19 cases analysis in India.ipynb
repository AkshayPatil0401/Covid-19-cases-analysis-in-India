```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
```

```python
df.head()
```

```python
df
```

```python
df.drop(['sno'], axis=1, inplace=True)
```

```python
df.tail()
```

```python
df.drop(df.index[36], axis=0, inplace=True)
```

```python
df.tail()
```

```python
total_states=np.arange(len(df['state_name']))
```

```python
total_states
```

```python
df['new_active']
```

# Total Positive cases bases on states in India

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(9,6), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states, df['positive'], align='edge', alpha=0.5,
        color=(0,0,1), 
        edgecolor=(0.8,0.8,0.7))
plt.yticks(total_states, df['state_name'])
plt.xlim(1, max(df['positive']+100))
plt.xlabel('Positive no. of cases')
plt.title('COVID-19 Total positive cases')
plt.show()
```

# Active no. of cases

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(9,6), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states, df['new_active'], align='edge', alpha=0.5,
        color=(1,0.5,0), 
        edgecolor=(0.8,0.8,0.7))
plt.yticks(total_states, df['state_name'])
plt.xlim(1, max(df['new_active']+100))
plt.xlabel('Active no. of cases')
plt.title('COVID-19 total Active no. of cases')
plt.show()
```

# total no. of new death cases 

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(9,6), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states, df['new_death'], align='edge', alpha=0.5,
        color=(1,0.5,0), 
        edgecolor=(0.8,0.8,0.7))
plt.yticks(total_states, df['state_name'])
plt.xlim(1, max(df['new_death']+100))
plt.xlabel('no. death of cases')
plt.title('COVID-19 total no. of death cases')
plt.show()
```
